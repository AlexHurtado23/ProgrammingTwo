## Selective Control Structures

Selective control structures are those that evaluate an expression, usually a Boolean condition, and from the result allow decisions to be made between one, two or more options. This is known as conditional selection.

There are 3 types of structures: simple selection (*If*), double selection (*If-else*) and multiple selection (*Else-if* and *Swich-Case*). The condition can use data of type integer, real or chararcter, and are generally formed using relational operators. For example X <Y or Z == 10.
___

### If 

The simple selective structure, better known as **If**, is the one that after evaluating an onition determines its value, which is true or false, and only if the result of the condition is true will the instruction or instructions defined by the condition be carried out. It is its sintax: 

if(condition 1){
instruction 1;
instruction 2;
instruction 3;
}

Where:

- condition 1: is the main condition to evalluate
- instruccion 1 - 3: are the following actions that the program will execute. 
- if the condition is false, the program will ingore the instruccions 1 - 3.
___

![If](https://8177home.files.wordpress.com/2018/11/if-statement.jpg)
___

**Example**

~~~
// Program to display a number if it is negative

#include <stdio.h>
int main() {
    int number;

    printf("Enter an integer: ");
    scanf("%d", &number);

    if (number < 0) {
        printf("You entered %d.\n", number);
    }
    
    printf("The if statement is easy.");
    
    return 0;
}
~~~

**Output 1**

-2 , You entered -2.  The if statement is easy.

**Output 2**

5  , The if statement is easy.

### If-else

The structure **if, else** is the one that allows evaluating a Boolean condition and choosing between 2 options. If the condition is true, it will be executed after the if, but if the condition is false, the instruction that follows the else will be executed. Therefore, the instructions are selected either after the If or after the else, but not both. Its syntax is the following:

if(condition 1){
instruccion 1;
instruccion 2;
}
else{
instruccion 3;
instruccion 4;
}

Where:

- condition 1: is the main condition to evaluate.
- instruccion 1 - 2: are the instruccions that will be executed if the condition is true.
- instruccion 3 - 4: are the instruccions that wiil be executed if the condition is false. 
___

![If else](https://8177home.files.wordpress.com/2018/11/if-else-statement.jpg)
___

**Example**

~~~
// Check whether an integer is odd or even

#include <stdio.h>
int main() {
    int number;
    printf("Enter an integer: ");
    scanf("%d", &number);

    // True if the remainder is 0
    if  (number%2 == 0) {
        printf("%d is an even integer.",number);
    }
    else {
        printf("%d is an odd integer.",number);
    }

    return 0;
}
~~~

**Output**

7 ,  7 is an odd integer.

### Conditional Operator ?

This operator **?** is used to express conditions in a program and can replace the ** if-else ** control structure. The operator requires a logical expression, which is evaluated and depending on whether it is true or false different instructions are executed, the syntax is:

<expression 1>? <expression 2>: <expression 3>

Where:

- <expression 1>: is the condition that will be evatuated (it most be boolean).
- <expression 2>: if <expression 1> is true, the program will execute it. 
- <expression 3>: if <expression 1> is false, the program will execute it.
___

### Else-if

Nested double selective, better known as **else-if** is one that within the scope of one condition has another condition; in other words, in an if - else another if - else is found.

Each condition will be evaluated in the order in which it appears: if condition 1 is true, the first instruction is executed and the other conditions are no longer reviewed. Otherwise the following condition is evaluated - 2 - and if this is true it will perform its instruction and so on. The final else containing the last statement will be executed if none of the above conditions was true. This structure is useful when you have 3 or more options. Its syntax is:

if(condition 1){instruccion 1;}
else if(condition 2){instruccion 2;}
     else if(condition 3) {instruccion 3;}
          else{instruccion 4;}
          
Where:

- condition 1: is the main contition to evalute.
- instruccion 1; is the instruccion that will be executed if the condition 1 is true.
- condition 2: is the second contition to evalute. if condition 1 is false.
- instruccion 2; is the instruccion that will be executed if the condition 2 is true.
- condition 3: is the third contition to evalute. if condition 2 is false.
- instruccion 3; is the instruccion that will be executed if the condition 3 is true.
- instruccion 4; is the instruccion that will be executed if the condition 3 is false.
___

![else if](https://8177home.files.wordpress.com/2018/11/if-else-if-statement.jpg)
___

**Example**

~~~
// Program to relate two integers using =, > or < symbol

#include <stdio.h>
int main() {
    int number1, number2;
    printf("Enter two integers: ");
    scanf("%d %d", &number1, &number2);

    //checks if the two integers are equal.
    if(number1 == number2) {
        printf("Result: %d = %d",number1,number2);
    }

    //checks if number1 is greater than number2.
    else if (number1 > number2) {
        printf("Result: %d > %d", number1, number2);
    }

    //checks if both test expressions are false
    else {
        printf("Result: %d < %d",number1, number2);
    }

    return 0;
}
~~~

**Output**

12 23 , Result: 12 < 23

### Swich-Case

Selective multiple or **Swich-Case** is that structure that allows choosing between 2 or more options, *swich* evaluates the expression that is inside the parentheses and the result is compared with the alternative values.

The *swich* in the expression implies the equal operator (==), so it compares if the expression is equal to any of the options. Therefore it cannot be compared using another relational operator. The data type of the expression can only be integer or character; therefore the options must match the data type of the expression.

*Swich* compares the value of the expression with each of the options in the order they are found. Each option is represented by the reserved word *case*, so there will be as many *case* as options on the program. Once it finds the equality of the expression with an option, the instructions will be carried out inside that option, until it finds a break. If it does not find any value equal to the expression, it will execute the *default* option if it exists.

The reserved word *break* should be used at the end of each case to interrupt the structure and not review the following options.The *default* will be the programmer's decision to include it, since it is optional. The syntax is:

swich(expresion)
{
    case 1: instruccion 1;
            break;
    case 2: instruccion 2;
            break;
    case n: instruccion n;
            break;
    default: instruction n+1;
}

Where:

- expression: is the integer or character that the user gives and the program must compare and find and option.
- case 1: is the first option.
- instruccion 1: is the instruccion that will be executed if the digit matchs with case 1. 
- case 2: is the second option.
- instruccion 2: is the instruccion that will be executed if the digit matchs with case 2.
- case n: is the nth option.
- instruccion n: is the instruccion that will be executed if the digit matchs with the nth case.
- default: is the last option.
- instruction n+1: is the instruccion that will be executed if any digit maches with any case. 
___

![Swich-Case](https://www.w3schools.in/wp-content/uploads/2014/06/php-switch.png)
___

**Example**

~~~
//This program compares the value of a variable and prints it. 

#include <stdio.h>
    int main() {
        int num = 8;
        switch (num) {
            case 7:
                printf("Value is 7");
                break;
            case 8:
                printf("Value is 8");
                break;
            case 9:
                printf("Value is 9");
                break;
            default:
                printf("Out of range");
                break;
        }
        return 0;
    }
~~~

**Output**

Value is 8

## Iterative Control Structures

An iterative structure allows repeating an action; repetition is controlled by an expression that is a boolean condition. There are three ways to express the loops: **for, while, and do/while**.

The body of the loop is constituted within the repetitive structure; repetition can be **Defined or Indefinite**.
The iteration is *defined* when the number of repetitions to be executed is known and the *indefinite* iteration when the number of times that an action will be repeated is not known.

--- 

### For-loop

The control structure **for** is generally used when repetition is defined. This structure handles all the details of repetition controlled by a counter, the representation of the structure is as follows:

for(expression 1;expression 2;expression 3){
    
  instruction
  
}

Where: 

- expression 1: It's the name of the variable and its start value.
- expression 2: It's the boolean expression.
- expression 3: Instruction to increase or decrease the control variable.
- instruction are the 'instructions' to be repeated.  

---

![loopfor](https://cdn.programiz.com/sites/tutorial2program/files/c-for-loop.jpg)

---

**Example:**

~~~
#include <stdio.h>
int main(){

    int i,suma=0;
    
    for(i=1;i<=100;i++){
    suma=suma+i;
    }
    
    printf("The sum is: %d\n",suma);
    
return 0;
}
~~~

**Output**

The sum is: 5050

---

### While-loop

In this structure, the repetition will be carried out as many times as indicated while a condition is met. The number of repetitions can be defined or indefinite. The representation of the structure is as follows:

expression 1;

while(expression2){

  instruction;
  
  expression 3;
  
}

Where:

- expression 1: It will always be the starting value of the control variable.
- expression 2: It's the boolean expression.
- expression 3: It's the loop counter. 
- instruction: It's the 'instruction' to execute.

---

![loopwhile](https://www.tutorialspoint.com/scala/images/scala_while_loop.jpg)

---

**Example:**

~~~
#include <stdio.h>
int main(){

    int i=1,suma=0;
    
    while (i<=150){
    suma=suma+i;
    i++;
    }
    
    printf("The sum is: %d\n",suma);
    
return 0;
}
~~~

**Output**

The sum is: 11325

---

### Do/while-loop

In this **do/while** structure, the loop continuation condition is tested at the end of the loop. It works similarly to the while structure; the difference is that one evaluates at the beginning of the cycle and the other at the end. In other words, this loop must perform the instruction at least once and then check if the condition is met. The representation of the structure is as follows:

expression 1;

do{

  instruction;
  
  expression 2;

} while(expression 3);

Where:

- expression 1: It's the start assignment.
- expression 2: It's the loop counter.
- expression 3: It's the boolean expression.
- instruction: It's the 'instruction' to execute.

---

![loopdowhile](https://www.tutorialspoint.com/scala/images/scala_do_while_loop.jpg)

---

**Example:**

~~~
#include <stdio.h>
int main(){

    int i,j;
    
    do{
        printf("Give me a number A: "); scanf("%d",&i);
        printf("Give me a number B: "); scanf("%d",&j);
      }
    while (i<j);
    
return 0;
}
~~~

**Output**

*This program will not stop until number A is not less than number B.*

### Break and Continue control statements


![break-continues](https://pediaa.com/wp-content/uploads/2018/10/Difference-Between-break-and-continue-Comparison-Summary.jpg)

---

**Example A: Program to calculate the sum of a maximum of 10 numbers, if a negative number is entered, the loop terminates**

~~~
#include <stdio.h>
int main(){
    int i;
    double number, sum = 0.0;

    for(i=1;i<=10;++i)
    {
        printf("Enter a n%d: ",i);
        scanf("%lf",&number);

        // If the user enters a negative number, the loop ends
        if(number < 0.0)
        {
            break;
        }

        sum += number; // sum = sum + number;
    }

    printf("Sum = %.2lf",sum);
    
return 0;
}
~~~

**Example B: Program to calculate the sum of a maximum of 10 numbers, negative numbers are skipped from the calculation**

~~~
#include <stdio.h>
int main(){

   int i;
   double number, sum = 0.0;

    for(i=1; i <= 10; ++i)
    {
        printf("Enter a n%d: ",i);
        scanf("%lf",&number);

        if(number < 0.0)
        {
            continue;
        }

        sum += number; // sum = sum + number;
    }

    printf("Sum = %.2lf",sum);
    
return 0;
}
~~~
