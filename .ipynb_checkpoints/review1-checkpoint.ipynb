## Operators

**Arithmetic operators:**
- (+)	addition or unary plus.
- (-)	subtraction or unary minus.
- (*)	multiplication.
- (/)	division.
- (%)	remainder after division.
- (++)  increment in 1.
- (--)  decrement in 1.

**Assignment operators:**
- **(=)**	a = b  Simple assignment operator. Assigns values from right side operands to left side operand
- **(+=)**	a = a+b Add AND assignment operator. It adds the right operand to the left operand and assign the result to the left operand.
- **(-=)**  a = a-b Subtract AND assignment operator. It subtracts the right operand from the left operand and assigns the result to the left operand
- **(*=)**	a = a*b Multiply AND assignment operator. It multiplies the right operand with the left operand and assigns the result to the left operand.
- **(/=)**	a = a/b 	Divide AND assignment operator. It divides the left operand with the right operand and assigns the result to the left operand
- **(%=)**	a = a%b Modulus AND assignment operator. It takes modulus using two operands and assigns the result to the left operand.

**Relational operators:**
- **(==)** Equal to.
- **(>)** Greater than.
- **(<)** Less than.
- **(!=)** Not equal to.
- **(>=)** Greater than or equal to.
- **(<=)** Less than or equal to.

**Logic operators:**
- **(&&)** Logical AND. True only if all operands are true.
- **(||)** Logical OR. True only if either one operand is true.
- **(!)** Logical NOT. True only if the operand is 0.

**Bitwise operators:**
- **(&)** Bitwise AND /Binary AND Operator copies a bit to the result if it exists in both operands.
- **(|)** Bitwise OR / Binary OR Operator copies a bit if it exists in either operand.
- **(^)** Bitwise exclusive OR /Binary XOR Operator copies the bit if it is set in one operand but not both.
- **(~)** Bitwise complement /Binary One's Complement Operator is unary and has the effect of 'flipping' bits.
- **(<<)** Shift left /Binary Left Shift Operator. The left operands value is moved left by the number of bits specified by the right operand.
- **(>>)** Shift right /Binary Right Shift Operator. The left operands value is moved right by the number of bits specified by the right operand.

**Strcuts operators:**
- **(->)**  perator in C allows to access elements in Structures and Unions.

## Control Structures

### Loops
 
- **for(){}** A for loop is a repetition control structure that allows you to efficiently write a loop that needs to execute a specific number of times
- **do{}while{}** If the test expression is true, the body of the loop is executed again and the test expression is evaluated.
- **while(){}** If the test expression is true, statements inside the body of while loop are executed. Then, the test expression is evaluated again.

### Conditionals 

- **if(){}** The statement is only executed if the condition is met.
- **else{}** The statement is only executed if the condition is not met.
- **elseif(){}** If one of them is fulfilled, the corresponding statement is executed and jumps to the end of the structure
- **"?: "** The conditional operator (? :) uses three operands and is the only ternary operator.
- **swich(){}** A switch statement allows a variable to be tested for equality against a list of values.
- **case{}** It is a test case If there is no match, the following default statements will be executed.
- **default** Is an argument to a function that a programmer is not required to specify


## Other reserved words 

- **break** The break statement ends the loop immediately when it is encountered.
- **continue** The continue statement skips the current iteration of the loop and continues with the next iteration.
- **return()** It returns a value outside the function. 

## Other functions

- **//** Single line comment.
- /**/ Area comment.
