# Types of Arrays in C
In c programming language, arrays are classified into two types. They are as follows:
*	One dimensional array
*	Multi dimensional array

## Single Dimensional Array
In the programming language c, one-dimensional arrays are used to store a list of values of the same data type, these can be of type integer string or string.
One-dimensional arrays are used to use one row of values.
In a one-dimensional matrix, the data is stored linearly. One-dimensional matrices are also called one-dimensional matrices, linear matrices, or simply 1-D matrices.
One-dimensional array declaration:
Example:
``` C
int rollNumbers [60];
char list [6];
int numbers[50]
``` 
___

### Initialization of Single Dimensional Array
We use the following general syntax for declaring and initializing a single dimensional array with size and initial values.

Example:

``` C
int list [6] = { 9, 0, 6, 8, 8, 8 };
```
___

## Multi Dimensional Array
An array is called as a multidimensional array when it has more than two dimensions.
In simple words, an array created with more than one dimension (size) is called a multidimensional array.
The multidimensional matrix can be two-dimensional or three-dimensional or four or more.

Most popular and commonly used multi dimensional array is two dimensional array. 
The 2-D arrays are used to store data in the form of table. We also use 2-D arrays to create mathematical matrices.

Example:
``` C
int matrix_A [2][3];
```
### Initialization of Two Dimensional Array

The following general syntax for declaring and initializing a two-dimensional array with a specified number of rows and columns with initial values.

Example:
``` C
int matrix_A [2][3] = { {1, 2, 3},{4, 5, 6} } ;
```
___ 
___

## Files

### Text File 

A **Text file** is a type of *digital*, *non-executable file* that contains letters, numbers, symbols and/or a combination. *It enables the creation and storage of text without any special formatting*. Text files are also known as **flat files** or **ASCII files**. 

A text file is used to *store standard and structured textual data or information that is human readable*. It is defined in several different formats, including the most popular **ASCII** for cross-platform usage, and **ANSI** for Windows-based operating platforms. It lacks *text formatting abilities*, such as *text alignment*, *bold text* and *font styles*.

**Text files characteristics**:

- **Content:** All content, including every character, will be preserved.
- **Context:** N/A. Plain Text does not contain metadata to describe the environment where it was originally created or used.
- **Behavior:** N/A. Plain Text does not provide interactive characteristics.
- **Structure:** The location of every linebreak will be preserved. There is no other predefined structure for Plain Text.
- **Appearance:** N/A. Plain Text does not store appearance attributes such as font (including font size or font style), color, alignment, etc. 

---
![textfile](https://excel.officetuts.net/wp-content/uploads/sites/2/2018/10/text-file.png) 
___


### Binary File

In c, a file is a logical concept that can be applied to many things from disk files to terminals or a printer. A sequence is associated with a specific file by performing a opening. Once the file is open, information can be exchanged between it and the program.


A **binary file** *is a sequence of bytes that have a one-to-one correspondence with a device
external*. So no character translation will take place. Also, the number of bytes written (read)
it will be the same as those found on the external device. Examples of these files are Photographs, images,
formatted text, executable files (applications), etc.

---
![binaryfile](https://upload.wikimedia.org/wikipedia/commons/thumb/7/76/Wikipedia_favicon_hexdump.svg/1200px-Wikipedia_favicon_hexdump.svg.png)
___

## File Managements

To work with the files in c we should follow a particular sequence or steps: 

1. Creation 
2. Opening 
3. Reading 
4. Writing
5. Close

**Check the following code:**

**Instruction:** *Write a program that reads a file and averages a student mark.*

``` C
#include<stdio.h>
#include<stdlib.h>
#include<string.h>

int main(){
    //We need to declarated the variable that we need for the following function.
    
    float num; //Numerical variable.
    char string[100]; //File content.
    float sum=0; // It will save each numerical item.
    float average=0; // it will save our average.
    
    //Creating a file
    
    /*
    We need to create a pointer variable to save our file and be able to access it.
    First we assign the value "NULL", but it is irrelevant because then we will assign the file, don't worry.
    */
    
    FILE *ptr_sub=NULL;
    FILE *ptr_ave=NULL;
    
    /*
    To assign a file and open it we use the function fopen(file_adress, mode).The address of the file is 
    placed in the first parameter or if we are creating a .txt file, only the name with its extension is placed, 
    for example: Source.txt, in this case we are using an already created file and the second parameter is the mode.
    To work a file we have different modes to do it:
    
    r: Open a text file for reading.
    w: Create a text file for writing.
    a: Open a text file to add.
    rb: Open a binary file for reading.
    wb: Create a binary file for writing.
    ab: Open a binary file to add.
    r+: Open a text file for read / write.
    w+: Create a text file for reading / writing.
    a+: Add or create a text file for read / write.
    */
    
    //Opening a file
    
    ptr_sub=fopen("C:\\Users\\USER\\Desktop\\UPY folder\\Subjects.txt","r"); //Here, we assign the file address.
    
    /*
    Here we create a new file, as a writing mode. Once created, 
    the next time the phrase will be rewritten.
    */
    
    ptr_ave=fopen("Average.txt","w");
    
    //we verify the variable has our file.
    
    if(ptr_sub==NULL){
        printf("Error opening the file.\n");
        return 1;
    }
    //we verify the variable has our file.
    if(ptr_ave==NULL){
        printf("Error opening the file.\n");
        return 1;
    }
       
    /*
    The following Loop-while will read our file and will save the content. We have 3 different functions to do it:
    
    fgetc () allows to read a character from the file, it receives a parameter and it is the variable of t
    he file. Returns the read character.
    
    fgets () reads entire strings from the file, until it finds a carriage return \n or the eof. 
    Three parameters accept the name of the character array where the read will be stored, the number of 
    bytes to read and the file variable. 
    
    fscanf () works in the same way as a traditional scanf () but with a 
    parameter at the beginning that indicates the file to read.
  
    */
    
    while(feof(ptr_sub)==0){ //It verify if it already reach the end.
        fscanf(ptr_sub,"%s%f",&string,&num); //We save our values.
        printf("%s %.2f\n",string,num); //We show it.
        sum=sum+num; //The sum of all numbers.
    }
    
    average=sum/8; // the sum divided between eight subjects
    
    To write something within other file we have 3 different functions to do it too. 
    
    fputc () writes a character to the file, it receives two parameters, the character to write and the file variable.
    
    fputs () writes a string to the file, it receives two parameters, the string to write, and the file variable.
    
    fprintf () works the same way as printf () but its first parameter is the file variable.
    */
    
    fprintf(ptr_ave,"The average is: %.2f\n",average); //We write the result within other file.
    
    // Closing a file
    
    //After working with the file, we close it.
    fclose(ptr_sub);
    fclose(ptr_ave);
    
    // Message to verify if the program run well
    printf("\n\nTask done");
  
return 0;
}
``` 
___

### Query

Query means that consulting information in a file, without make changes in the source file. This is mainly used for consulting information and data from servers.

**Example:**
*Program that reads a file and prints its content until the number of line given by the user*

``` C
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

const char FILE_data[]="data.txt";

int main(){

    int n_line=0; //Counter
    int num_line=0; //Current number of line
    char line[100]; //Current line
    FILE *data_ptr; //Pointer to data.txt
    
    data_ptr = fopen(FILE_data, "r"); //Opening data.txt for reading

/*Check if the file was well opened*/
	if (data_ptr == NULL) {
		fprintf(stderr, "Can not open data.txt");
		exit(8);
	}

/*Asking for the last line to read*/
    printf("Until what line you want to read?: ");
    scanf("%d",&num_line);
    printf("\n");
    
    /*Reading data.txt and printing  it until the given number of line*/
    while(!feof(data_ptr)){

        fgets(line, 100, data_ptr);

        if(!feof(data_ptr)){
            n_line++;

            if(n_line <= num_line){
                printf("%s", line);
            }
        }
    }
    fclose(data_ptr); //Closing data.txt
    printf("\n");
    printf("Reading complete! \n");

system("pause");
return 0;
}
``` 
---
**Content of input file**
- Line 1
- Line 2
- Line 3
- Line 4
- Line 5
- Line 6
- Line 7
- Line 8
- Line 9
- Line 10
---

**User gives**
- 5
---

**Output on the program**
- Line 1
- Line 2
- Line 3
- Line 4
- Line 5
___

### Modification 

Modify a file implies change its content, under this context we take a  source file called v1-file, the using a program we make changes on it, and we print the source file with the changes made in other file called v2-file.

**Example:**
*Program that reads a file, the asks for the new line and the number of line to be replaced in a new file*

``` C

#include<stdio.h>
#include<stdlib.h>
#include<string.h>

const char FILE_original[]="original.txt";
char original[100];

int main(){

    int num_line=0; //Number of line to be replaced
    int n_line=0; //Counter
    char line[100]; //Current line
    char new_line[100]; //New line
    FILE *original_ptr; //Pointer to original.txt
    FILE *new_ptr; //Pointer to new.txt

    original_ptr = fopen(FILE_original, "r"); //Opening original.txt for reading
	new_ptr = fopen("new.txt", "w"); //Creating and opening new.txt for writing
    
    /*Check if both files were well opened*/
	if (original_ptr == NULL) {
		fprintf(stderr, "Can not open original file %s\n", original);
		exit(8);
	}
	if (new_ptr == NULL) {
		fprintf(stderr, "Can not open new.txt");
		exit(8);
	}

/*Asking for a new line from the user*/
    printf("Input the new line: ");
    fgets(new_line, 100, stdin);

/*Asking fr the number of line in which the new line will be replaced*/
    printf("Number of line to be replaced: ");
    scanf("%d",&num_line);

/*Replacing the new line in new.txt*/
    while(!feof(original_ptr)){

        strcpy(line,"\0");
        fgets(line, 100, original_ptr);

        if(!feof(original_ptr)){
            n_line++;
          
            if(n_line != num_line){
                fprintf(new_ptr, "%s", line);
            }
            else{
                fprintf(new_ptr, "%s", new_line);
            }
        }
    }

    fclose(original_ptr); //Closing original.txt
    fclose(new_ptr); //Closing new.txt

    printf("\n");
    printf("The replacement of the given line was done, you can verify it \n");

system("pause");
return 0;
}
```
