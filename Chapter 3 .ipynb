## SUBPROGRAMS IN PROGRAMMING LANGUAGE

### WHAT'S IT?

Subprograms are procedures, routines or a set of instructions with the main purpose to realize a specific task within a certain general program.Those subprograms helps to divide the problem in small task where we can call them in any point of our general program and obtain new values to work. 

It is important to know that we can call subprograms within other suprograms. For example, We need to solve any programming problem and we have the main function and then we create another subprogram (subprogram A) but we realize that we need another subprogram with a relation to function A so, we create another subprogram (subprogram B) and we call it within the first subprogram (subprogram A). That is valid and that help you to organize and solve the problem in an easy way.

### Identifiers

1. An identifier declared in a block is accessible only from that block and all the blocks included in it (that block is considered **local**). A parameter formal is also considered a local declaration to the function block.

2. Identifiers declared outside any block are considered **global** and they can be used from any point in the program.


### Structure 

The basic subprogram structure is the following:

~~~
#include<stdio.h>

int z; // Global indetifier

int Sumar(int x, int y); // New subprogram Header

int main(){ //Main Fuction

 int suma;  //Local identifier
 z = 3;
 
 suma = Sumar(2, 3); // We call the subprogram
 
 printf("The sum is: %d",suma); // We print the result 
 
 return 0;
}

int Sumar(int x, int y) //Subprogram A
{

 z = x + y; // Operation
 
 return z; // It return the answer
 
}
~~~

### Exchange of Parameters by Value

1. The value of the real parameter is taken and assigned to the formal parameter. 
2. The formal parameter it is therefore a copy of the actual parameter. 
3. *Used exclusively for input parameters.*

~~~
#include<stdio.h>

int Value(int g, int h);

int main(){
    int a=100;
    int z=80;
    
    printf("The sum is: %d\n\n",Value(a,z));

return 0;
}

int Value(int g, int h){
return g+h;
}
~~~

### Exchange of Parameters by Reference 

1. Passing parameters by reference joins the variable of the formal parameter with the variable of the actual parameter and are treated as a single variable. They can only be passed by variables reference.
2. It is used for input / output parameters.
3. It differs from the step by value putting the symbol &.

~~~
#include<stdio.h>

void reference(int *k);

int main(){
    int a=10;
    reference(&a);
    printf("The New value is: %d\n\n",a);
return 0; 
}

void reference(int *k){
    *k+=5;
}
~~~

## Resursion

### What is it?

The process of calling a function by itself is called recursion and the function which calls itself is called recursive function. Recursion is used to solve various mathematical problems by dividing it into smaller problems. This method of solving a problem is called Divide and Conquer. 

In programming, it is used to divide complex problem into simpler ones and solving them individually

### Syntax of a recursive function

~~~
returntype recursive_func ([argument list])
{
    statements;
    ... ... ...
    recursive_func ([actual argument]);
    ... ... ...
}
~~~

### Warning !

Note: In order to prevent infinite recursive call, we need to define proper exit condition in a recursive function.

~~~
//Infinite Hello Word

#include<stdio.h>

int main()
{
    printf("Hello world");
    main();
    return 0;
}
~~~

In this program, we are calling main() from main() which is recursion. But we haven't defined any condition for the program to exit. Hence this code will print "Hello world" infinitely in the output screen.

### Direct

A function is said to be direct recursive if it calls itself directly.

~~~
//Program Function to show direct recursion

int fibo (int n)
{
    if (n==1 || n==2)
        return 1;
    else
        return (fibo(n-1)+fibo(n-2));
}
~~~

In this program, fibo() is a direct recursive function. This is because, inside fibo() function, there is a statement which calls fibo() function again directly.

### Indirect

A function is said to be indirect recursive if it calls another function and this new function calls the first calling function again.

~~~
//Program Function to show indirect recursion

int func1(int n)
{
    if (n<=1)
        return 1;
    else
        return func2(n);
}

int func2(int n)
{
    return func1(n);
}
~~~

In this program, func1() calls func2(), which is a new function. But this new function func2() calls the first calling function, func1(), again. This makes the above function an indirect recursive function.

### Another Example of recursion program

~~~
//Program to calculate factorial of a number using recursion.

#include<stdio.h>
int factorial(int n)
{
    if(n==0)
        return 1;
    else
        return (factorial(n-1)*n);
}

int main()
{
    int num,f;
    printf("Enter a number: ");
    scanf("%d",&num);
    f=factorial(num);
    printf("Factorial of %d = %d",num,f);
    return 0;
}
~~~

Here, factorial is calculated using recursion. The formula for calculating factorial of a number n is:

    n! = 1*2*...(n-1)*n 

Again, we can see

    (n-1)! = 1*2*...(n-1)

Hence we can write,

    n! = (n-1)! * n

We have implemented this recursive relation in our program.

Here:

- The number whose factorial is to be found is stored in the variable n.
- A recursive function factorial(num) calculates the factorial of the number.
- As factorial is (n-1)! * n, factorial function calculates the factorial by recursively multiplying n with factorial of (n-1).
- Finally, when n = 0, it returns 1 because 0! = 1.

**Output**

Enter a number: 7
Factorial of 7 = 5040


### Disadvantages of recursion

Recursive programs are generally slower than non recursive programs because it needs to make a function call so the program must save all its current state and retrieve them again later. This consumes more time making recursive programs slower.

Recursive programs requires more memory to hold intermediate states in a stack. Non recursive programs don't have any intermediate states, hence they don't require any extra memory.
