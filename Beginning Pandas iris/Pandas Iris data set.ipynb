```python
'''Step 1: Reading File'''

# Importing Pandas as pd
import pandas as pd

# Create a Data Frame by reading a file in the same folder. This file hasn't a header
df = pd.read_csv('iris.data', header=None) 

# Printing the whole Data Frame by wrting only its name: df

# Printing the head of the Data Frame 
df.head(6)
```

```python
'''Step 2: Changing name of columns'''

# Creating a vector with the names of each column
names = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Class']

# Assigning names to the columns by using a vector
df.columns = names

# Showing the tail of the Data Frame 
df.tail()
```

```python
'''Step 3: Transpose a Data Frame'''

# It makes columns into rows, and rows into columns
df.T
```

```python
'''Step 4: Ordered values inside columns'''

# It shows you the Data frame ordered by the given column, and its values ordered from highest to lowest
df.sort_values('Sepal_Length', ascending = False)

# If you want from lowest to highest: ascenfing = True
```

```python
'''Step 5: specific information of Data Frame'''

# It counts the values in a column 
df['Class'].value_counts()
```

```python
# It tells you the kind of data inside your Data Frame
df.dtypes
```

```python
# Showing specific columns
df[['Sepal_Length']]
```

```python
# Fist five rows 
df[:5]
```

```python
# Specific rows and columns
df.iloc[[5,6,7],[2,3]]  # df.iloc[[rows],[number of column]]
```

```python
# Filters
df[(df['Sepal_Length']>7)&(df['Sepal_Width']<3)]
```

```python
'''Step 6: NaN values'''

# Note: this data set hasn't NaN values, so let's create some

import numpy as np

df['Petal_Length'][:5]= np.nan

df.head(6)
```

```python
# Searching for NaN values
df.isna().sum()
```

```python
# Filling NaN values 
df['Petal_Length'] = df['Petal_Length'].fillna(0)
df.head(6)
```

```python
'''Step 7: Basic statistics of the Data Frame'''

df.shape # Gives the number of rows and columns of the file: output = (150, 5)

# It gives the basic statistical information about the Data Frame
df.describe()
```

```python
# Mean
df['Sepal_Width'].mean()
```

```python
# Mean
df_group = df.groupby('Class')['Sepal_Width'].mean()
df_group.name = 'mean_Sepal_Width'
df_group
```

```python
# Joining new column
df_join = df.join(df_group, on=('Class'), how= 'inner')
df_join.head(6)
```

```python
'''Step 8: Data Visualization'''

import matplotlib.pyplot as plt

# Printing a plot of the Data Frame
df_join.plot(figsize=[10,7])
```

```python
# Printing a column with an specific kind of plot 
df_join['Petal_Length'].plot(kind='bar')
```
